```python
#20 rounds of bagging
pred = {"pred00":np.array(np.repeat(np.nan,len(iris)))}
pred = pd.DataFrame(data=pred)
rounds = 50
models = []
for l in range(0,rounds):
    
    #Step 1 to get the bootstrap sample
    boot = iris.sample(len(iris),replace = True)
    
    # Step 2 get the indices for out-of-bag-data
    uniq_indx = np.unique(boot.index)
    test_index = []
    for m in range(0,len(iris)):
        if m not in uniq_indx:
            test_index.append(m)


    test = iris.iloc[test_index]
    #test.head()
    # Step 3 splitting X and Y 
    X_boot = boot.drop("Species",axis =1)
    Y_boot = boot["Species"]
    X_test  = test.drop("Species",axis =1)
    Y_test = test["Species"]
    
    # Step 4 Training the model on Bootstrap sample 
    tree = DecisionTreeClassifier(criterion = "gini",min_samples_leaf =5)
    tree.fit(X_boot,Y_boot)
    models.append(tree)
    
    # Step 5 get the predicted value on test data
    predict = tree.predict(X_test)
    
    # Step 6 add new predict column
    p = "pred"+str(l)
    pred[p] = pred["pred00"]
    
    # Step 7 bagging the second predicted value...
    j=0
    for n in range(0,len(iris)):
        if n in test_index:
            pred[p][n]=predict[j]
            j=j+1
        
pred
```
