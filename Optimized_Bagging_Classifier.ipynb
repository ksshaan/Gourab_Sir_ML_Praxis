```python
import numpy as np
import pandas as pd
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import confusion_matrix
```

```python
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
iris = iris.drop("Unnamed: 0",axis =1)
iris.head()
```

```python
#Creating a speparate Dataframe
pred = {"pred00":np.array(np.repeat(np.nan,len(iris)),dtype ="<U10")}
pred = pd.DataFrame(data=pred)
```

```python
#Bagging Round 1 
#Step 1 to get the bootstrap sample
boot = iris.sample(len(iris),replace = True)

# Step 2 get the indices for out-of-bag-data
uniq_indx = np.unique(boot.index)
test_index = []
for i in range(0,len(iris)):
    if i not in uniq_indx:
        test_index.append(i)


test = iris.iloc[test_index]
test.head()

# Step 3 splitting X and Y 
X_boot = boot.drop("Species",axis =1)
Y_boot = boot["Species"]
X_test  = test.drop("Species",axis =1)
Y_test = test["Species"]

# Step 4 Training the model on Bootstrap sample 
tree = DecisionTreeClassifier(criterion = "gini",min_samples_leaf =5)
tree.fit(X_boot,Y_boot)

# Step 5 get the predicted value on test data
predict = tree.predict(X_test)

# Step 6 add new predict column
pred["pred1"] = pred["pred00"]

# Step 6 bagging the second predicted value...
j=0
for i in range(0,len(iris)-1):
    if i in test_index:
        pred["pred1"][i]=predict[j]
        j=j+1
        
pred
```

```python
#20 rounds of bagging
pred = {"pred00":np.array(np.repeat(np.nan,len(iris)))}
pred = pd.DataFrame(data=pred)
rounds = 50
models = []
for l in range(0,rounds):
    
    #Step 1 to get the bootstrap sample
    boot = iris.sample(len(iris),replace = True)
    
    # Step 2 get the indices for out-of-bag-data
    uniq_indx = np.unique(boot.index)
    test_index = []
    for m in range(0,len(iris)):
        if m not in uniq_indx:
            test_index.append(m)


    test = iris.iloc[test_index]
    #test.head()
    # Step 3 splitting X and Y 
    X_boot = boot.drop("Species",axis =1)
    Y_boot = boot["Species"]
    X_test  = test.drop("Species",axis =1)
    Y_test = test["Species"]
    
    # Step 4 Training the model on Bootstrap sample 
    tree = DecisionTreeClassifier(criterion = "gini",min_samples_leaf =5)
    tree.fit(X_boot,Y_boot)
    models.append(tree)
    
    # Step 5 get the predicted value on test data
    predict = tree.predict(X_test)
    
    # Step 6 add new predict column
    p = "pred"+str(l)
    pred[p] = pred["pred00"]
    
    # Step 7 bagging the second predicted value...
    j=0
    for n in range(0,len(iris)):
        if n in test_index:
            pred[p][n]=predict[j]
            j=j+1
        
pred
```

```python
predictions = []
for i in range(0,pred.shape[0]):
    
    p = []
    p = pred.iloc[i,:]
    count = p.value_counts().index.tolist()
    predictions.append(count[0])
predictions
```

```python
cm = confusion_matrix(iris.Species,predictions)

(cm,(cm[0,0]+cm[1,1]+cm[2,2])/150)
```
