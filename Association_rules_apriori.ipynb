```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import apriori
```

```python
whiskey = pd.read_csv(r"C:\Users\hp\Documents\Praxis_Docs\GOURAB_SIR_PYTHON_ML\ML2\whiskey.csv")
```

```python
whiskey.head()
```

```python
whiskey.describe()
```

```python
freq_itemset = apriori(df=whiskey,min_support=0.1,use_colnames=True)
freq_itemset.sample(10)
```

```python
rules=association_rules(freq_itemset,min_threshold =0.5)
rules.sort_values(by='lift',ascending=False)

```
